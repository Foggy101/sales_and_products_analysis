# PRODUCT DEMAND FORECASTING

## <u>1. Preliminaries</u>

### 1.1 Importing packages

```python
#Import necessary packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

### 1.2 Loading data

```python
#Load the data
df_orders = pd.read_csv('data/orders.csv')
df_order_items = pd.read_csv('data/order_items.csv')
df_products = pd.read_csv('data/products.csv')
df_categories = pd.read_csv('data/product_category_name_translation.csv')
```

## <u>2. Data analysis</u>

### 2.1 Data preparation

```python
#Join the dataframes
merged_demand = pd.merge(df_orders, df_order_items, on='order_id', how='left')
merged_demand = pd.merge(merged_demand, df_products, on='product_id', how='left')
merged_demand = pd.merge(merged_demand, df_categories, on='product_category_name', how='left')

#Keep only selected columns
merged_demand = merged_demand[['order_id', 'order_status', 'order_purchase_timestamp', 'order_item_id', 'product_id', 'product_category_name_english']]

#Convert the order timestamp to just date
merged_demand['order_purchase_timestamp'] = pd.to_datetime(merged_demand['order_purchase_timestamp']).dt.date

#Drop canceled orders and the ones without product_ids
merged_demand = merged_demand[~merged_demand['product_id'].isna()]
merged_demand = merged_demand[merged_demand['order_status'] != 'canceled']

#Fill blank category names with 'no_information'
merged_demand['product_category_name_english'] = merged_demand['product_category_name_english'].fillna('no_information')

print('\nMERGED TABLE')
display(merged_demand.head())
print('\nDATA INFO')
display(merged_demand.info())
```

### 2.2 Time series analysis

```python
#Aggregate data
sales_data = (
    merged_demand
    .groupby(['order_purchase_timestamp', 'product_category_name_english', 'product_id'])
    .size()
    .reset_index(name='sales')
)

#Prepare time series data
sales_data = sales_data.set_index('order_purchase_timestamp')['sales']
sales_data.index = pd.to_datetime(sales_data.index)

#Resample and fill nulls
sales_data = sales_data.resample('D').sum().fillna(0)

#Decompose the time series
decomposition = seasonal_decompose(sales_data, model='additive')

#Break decomposition into subplots so it's possible to resize the graphs
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(16, 10))
decomposition.observed.plot(ax=ax1, title='Sales Over Time')
decomposition.trend.plot(ax=ax2, title='Trend Component')
decomposition.seasonal.plot(ax=ax3, title='Seasonal Component')

#Recreate original residual graph from statsmodels
ax4.set_xlim(sales_data.index.min(), sales_data.index.max())
ax4.plot(decomposition.resid.index.to_numpy(), decomposition.resid.to_numpy(), marker='o', linestyle='None')
ax4.set_title('Residual')
ax4.set_xlabel('order_purchase_timestamp')

#Display the results
plt.tight_layout()
plt.show()
```

### 2.3 Conclusions and model selection

After analyzing sales data and decomposing the time series both for all data and for separate categories it can be concluded that:
- There is a trend component to the series - overall sales seem to be increasing over time. However, the trend isn't significant enough to include it in a short-period forecasting model.
- There is a clear weekly seasonal component. Sales over the weekends are consitently higher than in weekdays. It must be included in the forecasting model.
- The residuals fluctuate around 0 with a roughly constant variance. It indicates that the decomposition has been successful - separating trend and seasonal components and leaving behind random noise with no apparent pattern.

Based on these points <u>SARIMA model is chosen</u> as it is exceptionally effective working with data exibiting a strong seasonal component and can work on a short period timeframe. <u>Random forest is chosen</u> as the second model. It's not as effective in handling big data as more robust models like LSTM, but should tackle a smaller dataset like this one very well.

#### Additional remarks regarding the time series:
- There is a significant drop in sales in the last 6 days of the series. Assumption can be made that this is due to a delay in obtaining transactional data from the source systems. It could also be related to various technical issues. Either way - for the sake of models accuracy last 6 days will be removed from the time series. The forecasting window will be expanded to compensate for that.
- The highest sales numbers can be observed in late November - clearly associated with increased demand around Black Friday. The SARIMA model will not capture this pattern.

## 3. <u>Model forecasting</u>

- <u>Forecast for all products in one category:</u> input <i>product_category_name_english</i> in <b>selected_category</b> and leave <b>selected_product</b> as an <i>empty string</i> and <b>new_product</b> as <i>False</i>.
- <u>Forecast for a single known product:</u> input <i>product_id</i> in <b>selected_product</b> and leave <b>selected_category</b> as an <i>empty string</i> and <b>new_product</b> as <i>False</i>.
- <u>Forecast for a single new product:</u> set <b>new_product</b> as <i>True</i>, input <i>product_category_name_english</i> in <b>selected_category</b> and insert <b>edtimated_share</b> of the category market.
    
    
<b>THEN RUN THE APPROPRIATE CELLS!</b>

### 3.1 User input

```python
#Set variables
selected_category = 'health_beauty'
selected_product = ''
new_product = False
estimated_share = 0.01 #Expected share of the categories market by the new product

forecast_period = 14 #Length of the forecasting period
ignored_days = 6 #Explained in 1.3 Additional remarks
forecast_lag = 7 #Lag in days before the demand will be calculated
forecast_days = forecast_period + forecast_lag
```

### 3.2 Data preparation 

```python
#Aggregate data
sales_data = (
    merged_demand
    .groupby(['order_purchase_timestamp', 'product_category_name_english', 'product_id'])
    .size()
    .reset_index(name='sales')
)

#Filter data by category or product
if new_product: #handle new products
    if selected_category != '':
        if selected_category in sales_data['product_category_name_english'].values:
            #products_in_category = len(sales_data[sales_data['product_category_name_english'] == selected_category])
            sales_data = sales_data[sales_data['product_category_name_english'] == selected_category]
        else:
            print('\nCategory not found')
    else:
        print('\nInput product category')
else: #handle existing products or categories
    estimated_share = 1
    if selected_category != '' and selected_product != '':
        print('\nInput either category or product - not both')
    else:
        if selected_category != '':
            if selected_category in sales_data['product_category_name_english'].values:
                sales_data = sales_data[sales_data['product_category_name_english'] == selected_category]
            else:
                print('\nCategory not found')

        if selected_product != '':
            if selected_product in sales_data['product_id'].values:
                sales_data = sales_data[sales_data['product_id'] == selected_product]
            else:
                print('\nProduct not found')

#Set index and resample
sales_data = sales_data.set_index('order_purchase_timestamp')['sales']
sales_data.index = pd.to_datetime(sales_data.index)
sales_data = sales_data.resample('D').sum().fillna(0)

#Create forecast time ranges
forecast_start_date = merged_demand['order_purchase_timestamp'].max() - pd.Timedelta(days=ignored_days - 1) #-1 so it starts from the next day from the available data
forecast_end_date = forecast_start_date + pd.Timedelta(days=forecast_days + ignored_days - 1)

#Resample the index including daily data from the whole data range
sales_data = sales_data.reindex(pd.date_range(sales_data.index.min(), forecast_start_date - pd.Timedelta(days=1)), fill_value=0)
```

### 3.3 SARIMA model forecasting

```python
#Set parameters
p = 1 #non-seasonal autoregressive order
d = 0 #non-seasonal differencing order, 0 because trend isn't considered in this short-period model
q = 1 #non-seasonal moving average order
P = 1 #seasonal autoregressive order
D = 1 #seasonal differencing order
Q = 1 #seasonal moving average order
s = 7 #length of a seasonal cycle

#Create the model
model = SARIMAX(sales_data * estimated_share, #share = 1 if new_product = False
                order=(p, d, q), 
                seasonal_order=(P, D, Q, s),
                enforce_stationarity=False,
                enforce_invertibility=False)

model_fit = model.fit(disp=False)

#Get forecasted values and confidence intervals
forecast = model_fit.get_forecast(steps=len(pd.date_range(forecast_start_date, forecast_end_date)))

forecast_demand = forecast.predicted_mean
forecast_demand_with_lag = forecast_demand[forecast_lag + ignored_days:]
forecast_ci = forecast.conf_int()

#Get forecasted demand over the forecast period
sum_demand = 0
if forecast_demand_with_lag.sum() < 0.1: #-1 because list index starts at 0, not 1
    sum_demand = 0
else:
    sum_demand = np.ceil(forecast_demand_with_lag.sum())
    
#Display results
print(f'\nFORECASTED DEMAND\n{forecast_demand_with_lag}')
print(f'\nEXPECTED DEMAND OVER THE FORECAST PERIOD: {sum_demand}')
print(f'(ACTUAL PREDICTION: {forecast_demand_with_lag.sum()})\n')

plt.figure(figsize=(16, 4))
if not new_product:
    plt.plot(sales_data, label='Observed Sales')
plt.plot(forecast_demand, label='Forecasted Demand', color='red')
plt.fill_between(forecast_ci.index, 
                 forecast_ci.iloc[:, 0], 
                 forecast_ci.iloc[:, 1], color='pink', alpha=0.3)
plt.title('Demand Forecast - SARIMA')
plt.xlabel('Date')
plt.ylabel('Demand')
plt.legend()
plt.show()
```

### 3.4 Random forest model forecasting

```python
# Suppress SettingWithCopyWarning
pd.options.mode.chained_assignment = None

#Prepare df for the model
df_sales = pd.DataFrame(sales_data * estimated_share) #share = 1 if new_product = False
df_sales.index.name = 'order_purchase_timestamp'

#Feature engineering
def create_features(df, label=None):
    df['date'] = df.index
    df['dayofweek'] = df['date'].dt.dayofweek
    df['quarter'] = df['date'].dt.quarter
    df['month'] = df['date'].dt.month
    df['year'] = df['date'].dt.year
    df['dayofyear'] = df['date'].dt.dayofyear
    df['dayofmonth'] = df['date'].dt.day
    df['weekofyear'] = df['date'].dt.isocalendar().week
    X = df[['dayofweek', 'quarter', 'month', 'year', 'dayofyear', 'dayofmonth', 'weekofyear']]
    if label:
        y = df[label]
        return X, y
    return X

X, y = create_features(df_sales, label='sales')

#Add lag features
for lag in range(1, 15):
    X[f'lag_{lag}'] = y.shift(lag)

#Train test split
X = X.dropna()
y = y[X.index]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

#Create the model
model_rf = RandomForestRegressor(n_estimators=100, random_state=42)
model_rf.fit(X_train, y_train)

#Make predictions
forecast_data = create_features(pd.DataFrame({'date': pd.date_range(forecast_start_date, forecast_end_date)}, index=pd.date_range(forecast_start_date, forecast_end_date)))
for lag in range(1, 15):
    forecast_data[f'lag_{lag}'] = df_sales['sales'].shift(lag)[-len(pd.date_range(forecast_start_date, forecast_end_date)):].values

forecast_demand = model_rf.predict(forecast_data.fillna(0))

#Calculate confidence intervals
forecast_ci_lower = forecast_demand - 1.96 * np.std(forecast_demand)
forecast_ci_upper = forecast_demand + 1.96 * np.std(forecast_demand)

#Get forecasted values within specified range
forecast_demand_with_lag = forecast_demand[forecast_lag + ignored_days:]
date_index_with_lag = pd.date_range(forecast_start_date, forecast_end_date)[forecast_lag + ignored_days:]
forecast_demand_with_lag = pd.Series(data=forecast_demand_with_lag, index=date_index_with_lag)

#Get forecasted demand over the forecast period
sum_demand = 0
if forecast_demand_with_lag.sum() < 0.1: #-1 because list index starts at 0, not 1
    sum_demand = 0
else:
    sum_demand = np.ceil(forecast_demand_with_lag.sum())
    
#Display results
print(f'\nFORECASTED DEMAND\n{forecast_demand_with_lag}')
print(f'\nEXPECTED DEMAND OVER THE FORECAST PERIOD: {sum_demand}')
print(f'(ACTUAL PREDICTION: {forecast_demand_with_lag.sum()})\n')

plt.figure(figsize=(16, 4))
if not new_product:
    plt.plot(sales_data, label='Observed Sales')

plt.plot(pd.date_range(forecast_start_date, forecast_end_date).to_numpy(), forecast_demand, label='Predicted Demand', color='red')
plt.fill_between(pd.date_range(forecast_start_date, forecast_end_date).to_numpy(), forecast_ci_lower, forecast_ci_upper, color='pink', alpha=0.3)

plt.title('Forecasted Demand - Random Forest')
plt.xlabel('Date')
plt.ylabel('Demand')
plt.legend()
plt.show()
```

# 4. <u>Model evaluation</u>

### 4.1 Calculations and plotting

```python
if not new_product:
    #Retrain the SARIMA model
    model_sarima = SARIMAX(sales_data[:-100] * estimated_share,
                    order=(p, d, q), 
                    seasonal_order=(P, D, Q, s),
                    enforce_stationarity=False,
                    enforce_invertibility=False)

    model_fit = model_sarima.fit(disp=False)
    forecast_sarima = model_fit.get_forecast(steps=100)
    forecast_sarima = forecast_sarima.predicted_mean

    #Get predictions for X_test and check its number of elements
    forecast_rf = model_rf.predict(X_test * estimated_share)
    prediction_range = len(forecast_rf[-100:]) #For some products with limited data available we won't have 100 data points in forecast_rf so we unify the amount of predictions across 2 models so we can compare them directly                

    #Display total demand across 100 days
    print(f'\nTOTAL DEMAND PREDICTIONS (OVER {prediction_range} DAYS)\nActual: {sales_data[-prediction_range:].sum():.2f}\nSARIMA prediction: {forecast_sarima[-prediction_range:].sum():.2f}\nRandom Forest prediction: {forecast_rf.sum()}\n')

    #Calculate MSE  
    mae_sarima = mean_absolute_error(sales_data[-prediction_range:], forecast_sarima[-prediction_range:])
    mae_rf = mean_absolute_error(sales_data[-prediction_range:], forecast_rf[-prediction_range:])

    print(f'\nMEAN ABSOLUTE ERROR (OVER {prediction_range} DAYS)\nMAE SARIMA: {mae_sarima},\nMAE Random Forest: {mae_rf}\n')

    #Display the results
    plt.figure(figsize=(16, 4))
    if not new_product:
        plt.plot(sales_data[-prediction_range:], label='Observed Sales')
    plt.plot(forecast_sarima[-prediction_range:], label='SARIMA Prediction', color='red', ls='dotted')
    plt.plot(np.array(y_test.index)[-prediction_range:], forecast_rf[-prediction_range:], label='Random Forest Prediction', color='green', ls='dashed')
    plt.title('Evaluation over last 100 days')
    plt.xlabel('Date')
    plt.ylabel('Demand')
    plt.legend()
    plt.show()
    
else:
    print("\nCan't evaluate the model for new products without actual data.")
```

### 4.2 Final conclusions

Two models were created and evaluated. They both can be used to forecast data for category and product demands. The recommended approach for existing products and categories would be to use both models and check their performance in section 4.1. Then either pick the model with better predictions / smaller mean absolute error or consider picking both and averaging their results. 

For new products without historical data we can't easily determine which model is better. Therefore, avereging their results is suggested. Still, the recommended approach would be to consult a domain expert to provide an existing product very similar to the new one and base the prediction of the new product demand on the historical data of the existing one.
